```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from scipy.spatial import Voronoi, voronoi_plot_2d

# ======================================================
# ASSOCIATIVE SELF-ORGANIZING MAP (A-SOM)
# ======================================================

class ASOM:
    def __init__(self, modalities, N=15, init="random", seed=None):
        if seed is not None:
            np.random.seed(seed)

        self.modalities = {k: np.asarray(v) for k, v in modalities.items()}
        self.mod_names = list(self.modalities.keys())
        self.N = N
        self.map_nodes = N * N

        # Shared 2D grid coordinates
        x, y = np.meshgrid(range(N), range(N))
        self.c = np.vstack([y.flatten(), x.flatten()]).T

        # SOM weights
        self.som_weights = {
            m: self._init_som(v, init)
            for m, v in self.modalities.items()
        }

        # Hebbian association matrices
        self.assoc = {}
        for i in range(len(self.mod_names)):
            for j in range(i + 1, len(self.mod_names)):
                a, b = self.mod_names[i], self.mod_names[j]
                self.assoc[(a, b)] = np.zeros((self.map_nodes, self.map_nodes))

    def _init_som(self, data, init):
        if init == "random":
            return np.random.rand(self.map_nodes, data.shape[1])
        raise ValueError("init must be 'random'")

    def _find_bmu(self, weights, vector):
        return np.argmin(np.linalg.norm(weights - vector, axis=1))

    # ---------------------
    # Training (Paper-Aligned)
    # ---------------------
    def train(self, epochs=3000, alpha_init=0.1):
        """
        Paper-matched hyperparameters:
        - Map size: 15x15
        - Initial learning rate: 0.1
        - Multiplicative decay: alpha *= 0.999
        - Sigma decays over the SAME number of iterations
        """

        n = len(next(iter(self.modalities.values())))
        alpha = alpha_init

        sigma_init = self.N / 2.0
        sigma_final = 1.0

        for t in range(epochs):
            i = np.random.randint(n)

            # Sigma decay (neighborhood influence radius)
            sigma = sigma_init * (1 - t / epochs) + sigma_final * (t / epochs)

            # BMUs for each modality
            bmus = {}
            for m in self.mod_names:
                sample = self.modalities[m][i]
                bmus[m] = self._find_bmu(self.som_weights[m], sample)

            # SOM weight updates
            for m in self.mod_names:
                sample = self.modalities[m][i]
                w = self.som_weights[m]

                d = np.linalg.norm(self.c - self.c[bmus[m]], axis=1)
                h = np.exp(-(d ** 2) / (2 * sigma ** 2))

                w += alpha * h[:, None] * (sample - w)

            # Hebbian association update
            for (a, b), matrix in self.assoc.items():
                matrix[bmus[a], bmus[b]] += 1

            # Learning rate decay (paper)
            alpha *= 0.999

        # Normalize association matrices
        for key in self.assoc:
            self.assoc[key] /= np.max(self.assoc[key])

    # ---------------------
    # Quantization Error
    # ---------------------
    def quantization_error(self):
        errs = []
        n = len(next(iter(self.modalities.values())))
        for i in range(n):
            for m in self.mod_names:
                sample = self.modalities[m][i]
                bmu = self._find_bmu(self.som_weights[m], sample)
                errs.append(np.linalg.norm(sample - self.som_weights[m][bmu]))
        return np.mean(errs)

    # ---------------------
    # Cross-Modal Generation
    # ---------------------
    def generate_cross(self, stim_mod, stim_vector):
        bmu_idx = self._find_bmu(self.som_weights[stim_mod], stim_vector)
        preds = {}
        for (a, b), matrix in self.assoc.items():
            if a == stim_mod:
                preds[b] = np.argmax(matrix[bmu_idx, :])
            elif b == stim_mod:
                preds[a] = np.argmax(matrix[:, bmu_idx])
        return preds

    # ---------------------
    # Visualizations
    # ---------------------
    def plot_som_3d(self, modality):
        W = self.som_weights[modality]
        fig = plt.figure(figsize=(7, 6))
        ax = fig.add_subplot(projection='3d')
        ax.scatter(W[:, 0], W[:, 1], W[:, 2], s=20)
        ax.set_title(f"SOM nodes ({modality})")
        plt.show()

    def plot_umatrix(self, modality):
        W = self.som_weights[modality].reshape(self.N, self.N, -1)
        umat = np.zeros((self.N, self.N))
        for i in range(self.N):
            for j in range(self.N):
                neighbors = []
                if i > 0: neighbors.append(W[i - 1, j])
                if i < self.N - 1: neighbors.append(W[i + 1, j])
                if j > 0: neighbors.append(W[i, j - 1])
                if j < self.N - 1: neighbors.append(W[i, j + 1])
                umat[i, j] = np.mean(
                    [np.linalg.norm(W[i, j] - n) for n in neighbors]
                )
        plt.figure(figsize=(6, 6))
        plt.imshow(umat, cmap='viridis')
        plt.colorbar()
        plt.title(f"U-matrix ({modality})")
        plt.show()

    def plot_activation_map(self, modality):
        hits = np.zeros(self.map_nodes)
        for x in self.modalities[modality]:
            hits[self._find_bmu(self.som_weights[modality], x)] += 1
        hits = hits.reshape(self.N, self.N)
        plt.figure(figsize=(6, 6))
        plt.imshow(hits, cmap='hot')
        plt.colorbar()
        plt.title(f"Activation map ({modality})")
        plt.show()

    def plot_cross_activation(self, stim_mod):
        hits = np.zeros(self.map_nodes)
        n = len(self.modalities[stim_mod])
        for i in range(n):
            sample = self.modalities[stim_mod][i]
            bmu_idx = self._find_bmu(self.som_weights[stim_mod], sample)
            hits[bmu_idx] += 1
        hits = hits.reshape(self.N, self.N)
        plt.figure(figsize=(6, 6))
        plt.imshow(hits, cmap='coolwarm')
        plt.colorbar()
        plt.title(f"Cross-modal activation ({stim_mod})")
        plt.show()

    def plot_bmu_centers(self, modality):
        centers = np.array([
            self._find_bmu(self.som_weights[modality], x)
            for x in self.modalities[modality]
        ])
        plt.figure(figsize=(6, 6))
        plt.scatter(self.c[:, 0], self.c[:, 1],
                    c='lightgray', s=50, label='Nodes')
        plt.scatter(self.c[centers, 0], self.c[centers, 1],
                    c='red', s=20, label='BMUs')
        plt.title(f'BMU Centers ({modality})')
        plt.legend()
        plt.show()

    def plot_bmu_voronoi(self, modality):
        centers = np.array([
            self._find_bmu(self.som_weights[modality], x)
            for x in self.modalities[modality]
        ])
        points = self.c[centers]
        if len(points) > 0:
            vor = Voronoi(points)
            voronoi_plot_2d(vor, show_vertices=False,
                            line_colors='orange', line_width=2)
            plt.scatter(self.c[:, 0], self.c[:, 1],
                        c='lightgray', s=30)
            plt.scatter(points[:, 0], points[:, 1],
                        c='red', s=20)
            plt.title(f'BMU Voronoi ({modality})')
            plt.show()

# ======================================================
# DEMO DATA (2 MODALITIES)
# ======================================================

X1, _ = make_swiss_roll(n_samples=2000)
X1 -= X1.mean(axis=0)
X2 = np.column_stack([
    np.sin(X1[:, 0]),
    np.cos(X1[:, 1]),
    X1[:, 2]
])

modalities = {
    "geometry": X1,
    "attributes": X2
}

# ======================================================
# TRAIN A-SOM
# ======================================================

msom = ASOM(modalities, N=15, init="random", seed=42)
msom.train(epochs=3000, alpha_init=0.1)

# ======================================================
# 1. Quantization Error
# ======================================================

qe_asom = msom.quantization_error()

control_som = {
    m: np.random.rand(msom.map_nodes, modalities[m].shape[1])
    for m in msom.mod_names
}

errs = []
for m in msom.mod_names:
    for x in modalities[m]:
        bmu = np.argmin(np.linalg.norm(control_som[m] - x, axis=1))
        errs.append(np.linalg.norm(x - control_som[m][bmu]))

qe_random = np.mean(errs)

print("Quantization Error (ASOM):", qe_asom)
print("Quantization Error (Random SOM):", qe_random)
print("Improvement factor:", qe_random / qe_asom)

# ======================================================
# 2. Hebbian Association Matrix
# ======================================================

a, b = 'geometry', 'attributes'
plt.imshow(msom.assoc[(a, b)], cmap='hot')
plt.colorbar()
plt.title("Hebbian Association Matrix (geometry  attributes)")
plt.show()

# ======================================================
# 3. Cross-Modal Generation Accuracy
# ======================================================

correct = 0
for i in range(len(modalities['geometry'])):
    stim = modalities['geometry'][i]
    pred = msom.generate_cross('geometry', stim)['attributes']
    true_bmu = msom._find_bmu(
        msom.som_weights['attributes'],
        modalities['attributes'][i]
    )
    if pred == true_bmu:
        correct += 1

accuracy = correct / len(modalities['geometry'])
chance = 1 / msom.map_nodes

print("Cross-modal generation accuracy:", accuracy)
print("Improvement factor over chance:", accuracy / chance)

# ======================================================
# 4. Visualizations
# ======================================================

msom.plot_cross_activation('geometry')
msom.plot_bmu_centers('geometry')
msom.plot_bmu_centers('attributes')
msom.plot_bmu_voronoi('geometry')
msom.plot_bmu_voronoi('attributes')
```
